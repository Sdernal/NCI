```python
import os
import xml.etree.cElementTree as ET
```

```python
files = os.listdir("./coropus_for_korzun/")
```

```python
output = open("./output.txt","w")
files.sort()

for fl in files:   
    XML_FILE = os.path.join("./coropus_for_korzun/", fl)
    tree = ET.ElementTree(file = XML_FILE)
    root = tree.getroot()
    kek = root.find("{http://www.abbyy.com/ns/Aux#}TextAnnotations").find("{http://www.abbyy.com/ns/Aux#}document_text")
    st = kek.text
    for s in st:    
        try:
            output.write(s)
        except UnicodeEncodeError as e:       
            print "Unicode Error in file: ", fl 
output.close()           
            
```
