```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
data_lstm4 = pd.read_csv('lstm4.csv')
index_lstm4 = np.array(data_lstm4.index)
fmTest_lstm4 = np.array(data_lstm4.fmTest)
fmTrain_lstm4 = np.array(data_lstm4.fmTrain)
acTest_lstm4 = np.array(data_lstm4.acTest)
acTrain_lstm4 = np.array(data_lstm4.acTrain)
```

```python
data_lstm12 = pd.read_csv('lstm12.csv')
index_lstm12 = np.array(data_lstm12.index)
fmTest_lstm12 = np.array(data_lstm12.fmTest)
fmTrain_lstm12 = np.array(data_lstm12.fmTrain)
acTest_lstm12 = np.array(data_lstm12.acTest)
acTrain_lstm12 = np.array(data_lstm12.acTrain)
```

```python
data_lstm8 = pd.read_csv('lstm8.csv')
index_lstm8 = np.array(data_lstm8.index)
fmTest_lstm8 = np.array(data_lstm8.fmTest)
fmTrain_lstm8 = np.array(data_lstm8.fmTrain)
acTest_lstm8 = np.array(data_lstm8.acTest)
acTrain_lstm8 = np.array(data_lstm8.acTrain)
```

```python
data_lstm8_2 = pd.read_csv('lstm8_2.csv')
index_lstm8_2 = np.array(data_lstm8_2.index)
fmTest_lstm8_2 = np.array(data_lstm8_2.fmTest)
fmTrain_lstm8_2 = np.array(data_lstm8_2.fmTrain)
acTest_lstm8_2 = np.array(data_lstm8_2.acTest)
acTrain_lstm8_2 = np.array(data_lstm8_2.acTrain)
```

```python
plt.figure(figsize=((16,8)))
plt.plot(index_lstm4,fmTest_lstm4,label="LSTM-4")
plt.plot(index_lstm12,fmTest_lstm12,label="LSTM-12")
plt.plot(index_lstm8,fmTest_lstm8,label="LSTM-8")
plt.plot(index_lstm8_2,fmTest_lstm8_2,label="LSTM-8_2")
plt.grid(True)
plt.legend(loc="lower right")
plt.show()
```

```python
plt.figure(figsize=((16,8)))
plt.plot(index_lstm4[100:],fmTest_lstm4[100:],label="LSTM-4")
plt.plot(index_lstm12[100:],fmTest_lstm12[100:],label="LSTM-12")
plt.plot(index_lstm8[100:],fmTest_lstm8[100:],label="LSTM-8")
plt.plot(index_lstm8_2[100:],fmTest_lstm8_2[100:],label="LSTM-8_2")
plt.grid(True)
plt.legend(loc="lower right")
plt.show()
```

```python
plt.figure(figsize=((16,8)))
plt.plot(index_lstm4,fmTrain_lstm4,label="LSTM-4")
plt.plot(index_lstm12,fmTrain_lstm12,label="LSTM-12")
plt.plot(index_lstm8,fmTrain_lstm8,label="LSTM-8")
plt.plot(index_lstm8_2,fmTrain_lstm8_2,label="LSTM-8_2")
plt.grid(True)
plt.legend(loc="lower right")
plt.show()
```

```python
plt.figure(figsize=((16,8)))
plt.plot(index_lstm4,acTest_lstm4,label="LSTM-4")
plt.plot(index_lstm12,acTest_lstm12,label="LSTM-12")
plt.plot(index_lstm8,acTest_lstm8,label="LSTM-8")
plt.plot(index_lstm8_2,acTest_lstm8_2,label="LSTM-8_2")
plt.grid(True)
plt.legend(loc="lower right")
plt.show()
```
